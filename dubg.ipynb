#CATBOOST

```
model_cat = CatBoostClassifier(od_type='Iter', iterations=10000, task_type='CPU')

model_cat.fit(X_train.astype(int), y_train.astype(int),
              eval_set=(X_test.astype(int), y_test.astype(int)),
              cat_features=X_train.columns[:-1],
              early_stopping_rounds=100)

labels_cat = model_cat.predict(X_test.astype(int))
print(f'\nScore: {accuracy_score(y_test, labels_cat)}')
```

```
skf_cat = StratifiedKFold(5)

probabilities_cat_ = []
probabilities_labels_cat_ = []
predictions_cat_ = []
accuracies_cat_ = []

i=1

skf_cat.get_n_splits(X,y)
for train_index, test_index in skf_cat.split(X,y):
  
  X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
  y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]

  model_cat_ = CatBoostClassifier(iterations=10000, od_type='Iter', task_type='CPU',learning_rate=0.1,depth=5)
  model_cat_.fit(X1_train.astype(int), y1_train.astype(int),
                 eval_set=(X1_test.astype(int), y1_test.astype(int)),
                 verbose=False, early_stopping_rounds=50,
                 cat_features=categorical_columns)
  acc = accuracy_score(y1_test, model_cat_.predict(X1_test.astype(int)))
  print(f'fold {i} : ')
  print(f'CatBoost 1: {acc}')
  probabilities_cat_.append(pd.DataFrame(data=model_cat_.predict_proba(test_df[X1_train.columns].astype(int))))
  predictions_cat_.append(pd.DataFrame(data=model_cat_.predict(test_df[X1_train.columns].astype(int)), index=test_df.ID, columns=['TARGET']))
  accuracies_cat_.append(acc)

  print('-'*30)
  i += 1

# print(np.mean(accuracies_cat))
print(np.mean(accuracies_cat_))
```

```
weights_cat = (accuracies_cat_/sum(accuracies_cat_))*10
# weights_lgbm = (accuracies_lgbm/sum(accuracies_lgbm))*10
# avg_prob_cat = sum(probabilities_cat_)/10
avg_prob_cat = sum([x*y for x,y in zip(probabilities_cat_, weights_cat)])/10
# avg_prob_lgbm = sum([x*y for x,y in zip(probabilities_lgbm, weights_lgbm)])/10
# avg_cat_lgbm = avg_prob_lgbm*0.25 + avg_prob_cat*0.75
avg_predictions = [list(row[1]).index(max(list(row[1]))) for row in avg_prob_cat.iterrows()]
avg_predictions = pd.DataFrame(data=avg_predictions, index=test_df.ID, columns=['TARGET'])
avg_predictions.TARGET = avg_predictions.TARGET.map(rev_map)
avg_predictions.to_csv('avg_submission_cat.csv')
avg_predictions
```
